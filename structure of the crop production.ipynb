```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the uploaded file (replace filename if needed)
df = pd.read_csv('produce.csv')

# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print("📋 Columns:", df.columns.tolist())

# Preview
df.head()
```

```python
# Look at unique values in 'particulars' to find rows related to "Production"
print("📌 Unique Particulars:\n", df['particulars'].unique())

# Filter rows that contain 'production' in 'particulars'
df_crop = df[df['particulars'].str.contains('production', case=False, na=False)]
print("✅ Filtered shape:", df_crop.shape)
df_crop.head()
```

```python
# Remove unnecessary columns
id_vars = ['particulars', 'unit', 'frequency']  # add 'state' if it exists

# Identify year columns automatically (those that look like '3-YYYY')
year_cols = [col for col in df_crop.columns if col.startswith('3-') and col[2:].isdigit()]
print("📆 Year columns:", year_cols)

# Melt the DataFrame (wide to long format)
df_long = df_crop.melt(id_vars=id_vars, value_vars=year_cols,
                       var_name='year', value_name='production')

# Clean year column
df_long['year'] = df_long['year'].str.extract(r'(\d{4})').astype(int)
df_long['production'] = pd.to_numeric(df_long['production'], errors='coerce')

# Drop missing production entries
df_long = df_long.dropna(subset=['production'])
print("🧾 Final reshaped data:\n", df_long.head())
```

```python
top_crops = df_long.groupby('particulars')['production'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_crops.values, y=top_crops.index, palette='viridis')
plt.title("🌾 Top 10 Crops by Total Production")
plt.xlabel("Total Production")
plt.ylabel("Crop Type")
plt.grid(True)
plt.show()
```

```python
yearly_total = df_long.groupby('year')['production'].sum()

plt.figure(figsize=(10,6))
sns.lineplot(x=yearly_total.index, y=yearly_total.values, marker='o')
plt.title("📈 Total Crop Production Over the Years")
plt.xlabel("Year")
plt.ylabel("Production")
plt.grid(True)
plt.show()
```
